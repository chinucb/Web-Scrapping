# use the requests library to download web pages

```python
import requests  
```

```python
topic_url='https://github.com/topics'
```

```python
response=requests.get(topic_url)
len(response.text)
```

```python
page_contents=response.text
```

```python
page_contents[:5000]
```

```python
with open('webpage.html', 'w', encoding='utf-8') as f:
    f.write(page_contents)
```

# use beautifulsoup to prase aand exstract information

```python
from bs4 import BeautifulSoup
```

```python
doc=BeautifulSoup(page_contents,'html.parser')
doc
```

```python
selection_class= "f3 lh-condensed mb-0 mt-1 Link--primary"
topic_titletags=doc.find_all('p',{'class':selection_class})  #('p',class_=selection_class)
len(topic_titletags)
```

```python
topic_titletags[:5]
```

```python
desc_selector="f5 color-fg-muted mb-0 mt-1"
topic_desc_tags=doc.find_all('p',{'class':desc_selector})
len(topic_desc_tags)
```

```python
topic_desc_tags[:5]
```

```python
topic_title_tag0=topic_titletags[0]
topic_title_tag0.parent.parent
```

```python
div=topic_title_tag0.parent
div
```

```python
topic_link_tags=doc.find_all('a',{'class':"no-underline flex-grow-0"})
topic_link_tags
```

```python
topic_heading="https://github.com" + topic_link_tags[0]['href']
print(topic_heading)
```

```python
topic_titles=[]
for tag in topic_titletags:
    topic_titles.append(tag.text)
print(topic_titles)
```

```python
topic_description=[]
for tag in topic_desc_tags:
    topic_description.append(tag.text.strip())
topic_description[:5]
```

```python
topic_urls=[]
base_url='https://github.com'
for tag in topic_link_tags:
    topic_urls.append(base_url + tag['href'])

topic_urls
    
```

 # create csv file(s)with the exstracted information

```python
import pandas as pd
```

```python
topic_dict ={
    'Title':topic_titles,
    'Description':topic_description,
    'URL': topic_urls
}
```

```python
topic_dict
```

```python
topic_df=pd.DataFrame(topic_dict)
topic_df
```

# to csv

```python
topic_df.to_csv('topic.csv')
```
